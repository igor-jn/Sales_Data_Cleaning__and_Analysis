## Not necessary anymore. Only for prepin TXT file

```python
file = open('C:\\Users\\NewHome32\\Desktop\\Scripts - Python\\My_Files\\BDCFHE - CASH Sales Audit Report by Dept Jan22.txt', 'r',  encoding="utf-8")
```

```python
read = file.readlines()
```

```python
no_header = read[11:]
```

```python
for line in no_header:
    for item in line:
        if item == ' ':
            line.remove(item)
```

```python
def clean(line):
    
    purg = line.split('  ')
    
    while '' in purg:
        purg.remove('')
        
    ','.join(purg)
        
    return purg
```

```python
write_to_file = []
```

```python
for line in no_header:
    
    write_to_file.append(clean(line))
    
write_to_file
```

```python
write_to_file[455:480]
```

## Writing new file

```python
with open("CASH Sales Audit Report by Dept Jan22.csv", "w", encoding='Utf-8') as f:
    writer = csv.writer(f)
    writer.writerows(write_to_file)
```
