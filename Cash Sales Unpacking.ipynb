## Cleaning function

```python
def clean(line):
    
    purg = line.split('  ')
    
    while '' in purg:
        purg.remove('')
        
    ','.join(purg)
        
    return purg
```

```python
write_to_file = []
```

```python
for line in no_header:
    
    write_to_file.append(clean(line))
```

## Writing new file

```python
with open("CASH Sales Audit Report by Dept 2022.csv", "w", encoding='Utf-8') as f:
    writer = csv.writer(f)
    writer.writerows(write_to_file)
```
